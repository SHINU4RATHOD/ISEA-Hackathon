# Split data in same ratio

```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Load your dataset
df = pd.read_csv(r"C:\Users\SHINU RATHOD\Desktop\ISEA Hackathon\clean_fraud_dataset_10000_Participants.csv")

# # Keep only relevant columns
# df = df[['message_text', 'attack_type']]

# # Relabel 'none' as 'benign'
# df['attack_type'] = df['attack_type'].replace('none', 'benign')

# Split: Train (70%) and Test (30%)
df_train, df_test = train_test_split(
    df,
    test_size=0.30,
    stratify=df['attack_type'],
    random_state=42
)

# Save the splits
df_train.to_csv("train.csv", index=False)
df_test.to_csv("test.csv", index=False)

# Display class distribution with ratio
def show_distribution(df, name):
    counts = df['attack_type'].value_counts()
    total = len(df)
    benign = counts.get("benign", 0)
    attacks = total - benign
    ratio = f"{benign}:{attacks}" if attacks != 0 else "N/A"
    print(f"{name} Set:")
    print(f"  Benign:     {benign}")
    print(f"  Attacks:    {attacks}")
    print(f"  Ratio:      {ratio}")
    print(f"  Breakdown:\n{counts}")
    print(f"  Total:      {total}\n")

show_distribution(df_train, "Train")
show_distribution(df_test, "Test")
```

```python
def ratio_format(df, name):
    counts = df['attack_type'].value_counts()
    benign = counts.get("benign", 0)
    attacks = sum(v for k, v in counts.items() if k != "benign")
    
    if attacks == 0:
        ratio = "∞ : 1"
    else:
        ratio_val = benign / attacks
        ratio = f"{ratio_val:.2f} : 1"
    
    print(f"{name} Ratio → {ratio}")

# Usage
ratio_format(df_train, "Train")
ratio_format(df_test, "Test")
```
